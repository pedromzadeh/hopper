```python
from pdf2image import convert_from_path
from glob import glob
import numpy as np
import matplotlib.pyplot as plt


def pdf_to_png(pdfs, dpi=300):
    if not isinstance(pdfs, list):
        pdfs = [pdfs]
    return [convert_from_path(p, dpi)[0] for p in pdfs]


def crop_to_size(png):
    w, h = png.size
    l, t = 310, 160
    return png.crop((l, t, w - l, h - t))
```

```python
# limit cycles
pdf_paths = glob("../output/IM/grid_id10/run_0/visuals/*")
numbers = [int(s.split("_")[-1].split(".")[0]) for s in pdf_paths]
pdf_paths = list(np.array(pdf_paths)[np.argsort(numbers)][68:86])
images = [crop_to_size(png) for png in pdf_to_png(pdf_paths)]

fig, axs = plt.subplots(1, len(images), figsize=(2 * len(images), 5))
for ax, k in zip(axs, images):
    ax.imshow(k)
    ax.axis("off")

axs[0].text(0, -20, r"$\Delta_t=$" + "10 min" + r"$\rightarrow$", fontsize=40)
# axs[0].text(-150, -20, "(i)", fontsize=40)
axs[0].annotate(r"$x$", xy=(0, 0), xytext=(-90, 50), fontsize=30)
axs[0].annotate(r"$-0$", xy=(0, 0), xytext=(-73, 250), fontsize=30)
axs[0].annotate(
    "",
    xy=(-30, 10),
    xycoords= "axes points",
    xytext=(-30, 270),
    fontsize=40,
    arrowprops=dict(facecolor="black", lw=2, arrowstyle="-|>"),
)
plt.subplots_adjust(wspace=-0.2)
plt.savefig("paper_plots/limit_cycle_time_series.pdf")
```
